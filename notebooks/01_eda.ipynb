# Exploratory Data Analysis — Legal Case Outcomes

Inspecting class distribution, text length, and vocabulary patterns before modeling.

```python
import pandas as pd
import matplotlib.pyplot as plt
from src.data.preprocess import load_raw, run as preprocess, TEXT_COL, TARGET

plt.rcParams['figure.figsize'] = (10, 5)
```

## 1. Load and Clean

```python
raw = load_raw()
df = preprocess(raw)
df.head()
```

## 2. Class Distribution

Legal outcome categories are naturally imbalanced — some outcomes (e.g. 'cited', 'applied') are far more common than rare ones (e.g. 'overruled'). This shapes how we handle class weighting in both models.

```python
df[TARGET].value_counts().plot(kind='bar', color='steelblue', title='Case Outcome Distribution')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 3. Text Length by Outcome

```python
df['text_length'] = df[TEXT_COL].str.split().apply(len)
df.boxplot(column='text_length', by=TARGET, rot=45, figsize=(12, 6))
plt.title('Case Text Length by Outcome')
plt.ylabel('Word Count')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

## 4. Most Frequent Terms (Overall)

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=20, stop_words='english')
term_matrix = vectorizer.fit_transform(df[TEXT_COL].sample(5000, random_state=42))
term_counts = term_matrix.sum(axis=0).A1
terms = vectorizer.get_feature_names_out()

pd.Series(term_counts, index=terms).sort_values().plot(kind='barh', color='coral', title='Top 20 Terms')
plt.tight_layout()
plt.show()
```
